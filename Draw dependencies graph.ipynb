```python
arrofedgelist = []
keyarr = list(p.keys())
#loop through every dependency arr
for i in range(len(keyarr)):
    curlist = p.get(keyarr[i])
    #curkey
    curkey = keyarr[i]
    for j in range(len(curlist)):
        curval = curlist[j]
        #put into tuple
        tmptuple = (curval,curkey)
        arrofedgelist.append(tmptuple)
```

```python
arrofedgelist
```

```python

G = nx.MultiDiGraph()
G.add_edges_from(arrofedgelist)
# pos = nx.spring_layout(G)
# nx.draw_networkx_edge_labels(G)
nx.draw(G,with_labels = True)
```

```python
with open('./edge_list.csv','w') as out:
    csv_out=csv.writer(out)
    for tup in arrofedgelist:
        csv_out.writerow(tup)
```
